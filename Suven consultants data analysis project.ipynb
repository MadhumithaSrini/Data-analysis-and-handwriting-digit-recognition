```python
#importing all the packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#reading the csv file and finding out the names of the columns
df = pd.read_csv("C:/Users/madhumitha/Downloads/weatherHistory.csv")
#df.head()
df.columns
```

```python
df.head()
```

```python
#It is a general step to analyse how many null values are there in a column
df.isnull().sum()
```

```python
# the date column has been converted into datetime format
import datetime
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df.dtypes
```

```python
df.head()
```

```python
#azzessing the year, month, hour, secs as we have converted the date column into datetime format
df['year'] = pd.DatetimeIndex(df['Formatted Date']).year
df['month'] = pd.DatetimeIndex(df['Formatted Date']).month
df['date'] = pd.DatetimeIndex(df['Formatted Date']).date
df['hour'] = pd.DatetimeIndex(df['Formatted Date']).hour
df['minute'] = pd.DatetimeIndex(df['Formatted Date']).minute
```

```python
df.head()
```

```python
#sorting the values in the order of the date and then hour and min for easy understanding and processing of the data
df = df.sort_values(by = ["date","hour","minute"], axis=0, ascending=True, inplace=False)
```
